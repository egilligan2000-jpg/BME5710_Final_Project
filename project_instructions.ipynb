### Instructions
In this project, you will develop a CNN-based MRI inpainting method using PyTorch. Inpainting is the problem of recovering missing pixels from a partially observed image - a common challenge in medical imaging when data acquisition is incomplete or corrupted.

* From the $m_1$ training images, carve out a validation split for hyperparameter tuning. A reasonable split is 225 images for training and 25 for validation, but you are free to adjust this.
* Each training sample consists of a full MRI image $\boldsymbol{y}_{\sf tr}^{(i)}$ and a randomly generated binary mask $\boldsymbol{m}^{(i)}$ (75% pixels missing). The masked image $\boldsymbol{x}^{(i)} = \boldsymbol{m}^{(i)} \odot \boldsymbol{y}_{\sf tr}^{(i)}$ serves as the network input (along with the mask itself), and the full image $\boldsymbol{y}_{\sf tr}^{(i)}$ serves as the target output. The masks should be regenerated randomly each epoch during training.
* Train your CNN to minimize a loss function, e.g., mean square error, between the ground-truth images $\{\boldsymbol{y}_{\sf tr}^{(i)}\}_{i=1}^{m_1}$ and the reconstructed images $\{\widehat{\boldsymbol{y}}_{\sf tr}^{(i)}\}_{i=1}^{m_1}$.
* Use the validation images to fine-tune your network. Do not use these images to train the network.
* Once trained, apply your network to the $m_3 = 50$ masked test images in `test-masked/`. Each test image comes with a fixed binary mask — use the provided mask as the second input channel to the network. Generate a Kaggle submission file (`submission.csv`) using the helper code in `demo_inpainting.ipynb`.
* Kaggle will score your submission using the competition metric $\frac{1}{m_3}\sum_{i=1}^{m_3}\big[{\sf PSNR}(\boldsymbol{y}_{\sf te}^{(i)}, \widehat{\boldsymbol{y}}_{\sf te}^{(i)}) + 50 \times {\sf SSIM}(\boldsymbol{y}_{\sf te}^{(i)}, \widehat{\boldsymbol{y}}_{\sf te}^{(i)})\big]$, where ${\sf PSNR}(\boldsymbol{y}, \widehat{\boldsymbol{y}}) = 20\log_{10}\big[{\max(\boldsymbol{y})}/{\sqrt{\frac{1}{N}\|\boldsymbol{y} - \widehat{\boldsymbol{y}}\|_2^2}}\big]$. Since you do not have access to the test ground truth, optimize this metric on your validation fold as a proxy.
* You are allowed up to **five Kaggle submissions per day** per team. Coordinate with your teammates to use these attempts wisely.

### Mindset while training a DL model
To train your DL model, you will try many different options and see what sticks. This process can be exciting and frustrating at the same time. On one hand, you will see your DL tool improve incrementally, but on the other hand, you may find the hyperparameter tuning process frustrating. A bit of frustration is expected, but this exploration does not have to be a directionless, random process. Use intuition and leverage your understanding of basic ML concepts (e.g., over- and underfitting) to navigate. Also, read scientific literature and blogs to see what others have to say about a specific design choice.
* Be creative! Think outside the box! Don't be afraid to try some "crazy" ideas.
* Experiment with different loss functions. Sometimes a composite loss function, e.g., one based on $\ell_1$ loss and SSIM, can outperform a loss function based on just MSE or $\ell_1$.
* Experiment with the overall architecture of the CNN. UNet and ResNet are two popular choices for inpainting, but they are not the only ones.
* Try different activation functions, perhaps something other than ReLU or leaky ReLU.
* The number of layers, number of channels, size of kernels, batch normalization, dropout, data augmentation, transfer learning, learning rate, number of epochs, batch size, optimizer, regularization, skip connections, and learning rate scheduling can all have an impact on the performance of a network. The possibilities are endless :-)
* For inpainting specifically, how the network uses the mask as an input can matter a great deal. Think carefully about whether the network should treat observed and missing pixels differently.
